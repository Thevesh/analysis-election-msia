```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

voters_ge14 = 14940624
voters_ge15 = 21173638

data = [
    ["08:00",np.nan,0.0001],
    ["09:00",0.0001,9],
    ["10:00",24,20],
    ["11:00",36,31],
    ["12:00",47,42],
    ["13:00",55,50],
    ["14:00",62,58],
    ["15:00",69,65],
    ["16:00",76,70],
    ["17:00",82.32,72],
    ["18:00",np.nan,73.11],
]

df = pd.DataFrame(data=data,columns=['time','ge14','ge15'])

df = pd.read_csv('data-local/turnout.csv').set_index('time')
df['voters_ge14'] = (df.ge14*voters_ge14/1E8).round(1)
df['voters_ge15'] = (df.ge15*voters_ge15/1E8).round(1)
df.columns = ['GE14 Turnout','GE15 Turnout','GE14 Votes','GE15 Votes']

plt.rcParams.update({'font.size': 11,
                     'font.family': 'sans-serif',
                     'grid.linestyle': 'dashed'})
plt.rcParams["figure.figsize"] = [8,12]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(2,1)

axes = ax.ravel()

df.plot(kind='bar', width=0.7, y=[x for x in df.columns if 'Votes' in x], edgecolor='black', lw=0, color=['#D0D0D0','red'], ax=axes[0])
axes[0].set_title('GE15 Turnout Chart (@thevesh on Twitter for charts; data from @sprgovmy)\n\nNumber of Votes Cast (millons): GE15 vs GE14\n')

df.plot(kind='bar', width=0.7, y=[x for x in df.columns if 'Turn' in x], edgecolor='black', lw=0, color=['#C0C0C0','#010027'], ax=axes[1])
axes[1].set_title('\nVoter Turnout (% of registered voters): GE15 vs GE14\n')

for i in [0,1]:
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)
    axes[i].yaxis.grid(True)
    axes[i].set_axisbelow(True)
    axes[i].set_xlabel('')
    axes[i].legend(labelspacing = 1.5)
    for c in axes[i].containers:
        if i == 0: labels = [f'{h:0.1f}' if (h := v.get_height()) != 0 else '' for v in c ]
        if i == 1: labels = [f'{h:0.0f}' if (h := v.get_height()) != 0 else '' for v in c ]
        axes[i].bar_label(c, labels=labels,fontsize=10)
plt.suptitle('\nNote: In GE15, we may see higher votes despite a lower turnout %\n\nThe voter roll has grown by 41% (from 14.9 to 21.2 mil) due to AVR and Undi-18', y=0)

# code to generate ALT text

str = (f'A bar chart contrasting the cumulative number of voters and voter turnout in GE15 against GE14, updated hourly.\n')

for i in range(1, len(df)):
    str += \
    (f'\nAs of {df.index[i]}: '
    f'{df["GE15 Turnout"][i]}% turnout with '
    f'{df["GE15 Votes"][i]} mil votes today, versus '
    f'{df["GE14 Turnout"][i]}% turnout with '
    f'{df["GE15 Votes"][i]} mil votes at GE14')
str = str.replace('.0%','%').replace('nan%','?%')
print(str)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = [
    ["08:00",np.nan,0.0001,0,0],
    ["09:00",0.0001,9,9,9],
    ["10:00",24,20,18,19],
    ["11:00",36,31,27,28],
    ["12:00",47,42,35,36],
    ["13:00",55,50,41,43],
    ["14:00",62,58,45,49],
    ["15:00",69,65,50,57],
    ["16:00",76,70,54,63],
    ["17:00",82.32,72,56,66],
    ["18:00",np.nan,73.11,58,69]
]

df = pd.DataFrame(data=data,columns=['time','ge14','ge15','tioman','pserai']).set_index('time')
df.columns = ['GE14 Turnout','GE15 (19 Nov)','N.42 Tioman','P.017 Padang Serai']
for c in df.columns: df[c] = df[c].round(0)

plt.rcParams.update({'font.size': 11,
                     'font.family': 'sans-serif',
                     'grid.linestyle': 'dashed'})
plt.rcParams["figure.figsize"] = [8.8,6]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()


df.plot(kind='bar', width=0.7, y=[x for x in df.columns if 'GE14' not in x], edgecolor='black', lw=0, color=['#C0C0C0','#6fbcf0','#010027'], ax=ax)
ax.set_title('\nVoter Turnout:  P.017 Padang Serai & N.42 Tioman\n\ncharts: @thevesh on Twitter\n\ndata: @sprgovmy')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True)
ax.set_axisbelow(True)
ax.set_xlabel('')
ax.legend(labelspacing = 1.5)
for c in ax.containers:
    labels = [f'{h:0.0f}' if (h := v.get_height()) != 0 else '' for v in c ]
    ax.bar_label(c, labels=labels,fontsize=10)

str = (f'A bar chart showing the voter turnout in the two delayed elections '
f'being held in P.017 Padang Serai and N.42 Tioman today,' 
f'with the general GE15 turnout from 19th November for comparison.\n')

for i in range(1, len(df)):
    str += \
    (f'\nAs of {df.index[i]}: '
    f'{df["P.017 Padang Serai"][i]}% in Padang Serai and '
    f'{df["N.42 Tioman"][i]}% in Tioman, vs '
    f'{df["GE15 (19 Nov)"][i]}% in GE15')
str = str.replace('.0%','%').replace('nan%','?%')
print(str)
```
